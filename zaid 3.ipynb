```python
import numpy
import pandas
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import keras
from keras import layers

```

```python
# load dataset
dataframe = pandas.read_csv("housing.csv", delim_whitespace=True, header=None)
dataset = dataframe.values
# split into input (X) and output (Y) variables
X = dataset[:,0:13]
Y = dataset[:,13]
```

```python
# define base model
def baseline_model():
    # create model
    inputs = keras.Input(shape=(13,))
    x=layers.Dense(60, activation='relu')(inputs)
    x=layers.Dense(30, activation='relu')(x)
    outputs = layers.Dense(1)(x)


    # Compile model
    model = keras.Model(inputs , outputs)
    model.compile(metrics=['accuracy'] , optimizer = 'Adam' , loss='binary_crossentropy')
    
    return model

```

```python
# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
# evaluate model with standardized dataset
estimator = KerasRegressor(build_fn=baseline_model, epochs=100, batch_size=5, verbose=0)
```

```python
kfold = KFold(n_splits=10, random_state=seed)
results = cross_val_score(estimator, X, Y, cv=kfold)
print("Results: %.2f (%.2f) MSE" % (results.mean(), results.std()))
```

```python
# evaluate model with standardized dataset
numpy.random.seed(seed)
estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasRegressor(build_fn=baseline_model, epochs=50, batch_size=5, verbose=0)))
pipeline = Pipeline(estimators)
kfold = KFold(n_splits=10, random_state=seed)
results = cross_val_score(pipeline, X, Y, cv=kfold)
print("Standardized: %.2f (%.2f) MSE" % (results.mean(), results.std()))
```

# Step 8. Rewriting the code by doing Model Subclassing

```python
def larger_model():
    
    class MyModel(keras.Model):
        def __init__(self):
            super(MyModel, self).__init__()
            self.dense1 = layers.Dense(60, activation = 'relu')
            self.dense2 = layers.Dense(30 , activation = 'relu')
            self.dense3 = layers.Dense(10 , activation = 'relu')
            self.dense4 = layers.Dense(1)
        
        def call(self , inputs):
            x = self.dense1(inputs)
            x = self.dense2(x)
            x = self.dense3(x)
            return self.dense4(x)
    


    model = MyModel()
    model.compile(metrics=['accuracy'] , loss = 'mse' , optimizer = 'Adam')
    return model
    
```

```python
# evaluate model with standardized dataset
numpy.random.seed(seed)
estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasRegressor(build_fn=larger_model, epochs=50, batch_size=5, verbose=1)))
pipeline = Pipeline(estimators)
kfold = KFold(n_splits=10, random_state=seed)
results = cross_val_score(pipeline, X, Y, cv=kfold)
print("Standardized: %.2f (%.2f) MSE" % (results.mean(), results.std()))
```
